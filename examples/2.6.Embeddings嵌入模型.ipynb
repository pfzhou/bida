#### 2.6 Embeddings嵌入模型


- OpenAI

```python
from bida import Embedding

em = Embedding(model_type='openai')

result = em.create("hello world")
print(f"返回：{len(result)}个{len(result[0])}维的embedding，内容是：\n{result}")
```

```python
from bida import Embedding

em = Embedding(model_type='openai')
strlist = ["hello world", "hello china", " ", " \n \n\n", "hello AI"]
result = em.create(strlist)
print(f"返回：{len(result)}个{len(result[0])}维的embedding，内容是：\n{result}")
```
